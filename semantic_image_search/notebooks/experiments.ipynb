```python
# Get existing collections
collections = qdrant_client.get_collections().collections
existing_names = {c.name for c in collections}

# Check and create
if COLLECTION_NAME not in existing_names:
    print(f"Creating collection: {COLLECTION_NAME}")
    qdrant_client.create_collection(
        collection_name=COLLECTION_NAME,
        vectors_config=models.VectorParams(
            size=VECTOR_SIZE,
            distance=models.Distance.COSINE,
        ),
    )
    print(f"Created new collection: {COLLECTION_NAME}")
else:
    print(f"Collection already exists: {COLLECTION_NAME} (reusing)")
```

```python
import os
import numpy as np
from PIL import Image
from uuid import uuid4, uuid
```

```python
import uuid

# Root embeddings function
def index_image(image_path, category=None):
    image_path = Path(image_path)

    # Create embedding
    img_embed = embedder.embed_image([str(image_path)])[0]
    emb = np.array(img_embed).tolist()

    payload = {
        "filename": image_path.name,
        "path": str(image_path),
        "category": category,
    }

    point_id = str(uuid.uuid4())

    qdrant_client.upsert(
        collection_name=COLLECTION_NAME,
        points=[
            models.PointStruct(
                id=point_id,
                vector=emb,
                payload=payload,
            )
        ]
    )
    print(f"‚úÖ Indexed: {image_path.name}  | Category: {category}")
```

```python
# Root folder indexing
def index_folder(root_folder):
    root_path = Path(root_folder)

    if not root_path.exists():
        raise FileNotFoundError(f"Folder does not exist: {root_path}")

    indexed = 0
    
    # Walk every file in every subfolder
    for file in root_path.rglob("*"):
        if file.is_file() and file.suffix.lower() in [".jpg", ".jpeg", ".png", ".webp", ".bmp"]:
            
            category = file.parent.name  # infer category
            
            index_image(str(file), category=category)
            indexed += 1

    print(f"\nüåü DONE ‚Äî Indexed {indexed} images from: {root_path}")
```

```python
def search_by_text(query: str, k: int = 5):
    # 1. Get text embedding
    query_vec = embedder.embed_query(query)
    query_vec = np.array(query_vec).tolist()

    # 2. Use query_points for retrieval
    results = qdrant_client.query_points(
        collection_name=COLLECTION_NAME,
        query=query_vec,
        limit=k,
        with_payload=True
    )

    print(f"\nüîé Text search ‚Üí \"{query}\"")

    hits = []
    for i, r in enumerate(results.points, start=1):
        payload = r.payload or {}
        print(
            f"{i}. {payload.get('filename')} "
            f"(category={payload.get('category')}, score={r.score:.4f})"
        )
        hits.append({
            "id": r.id,
            "score": r.score,
            "filename": payload.get("filename"),
            "path": payload.get("path"),
            "category": payload.get("category"),
        })

    return hits
```

```python
search_by_text("image of cat with angry face", k=5)
```

```python
def search_by_image(image_path, k: int = 5):
    image_path = Path(image_path)

    img_vec = embedder.embed_image([str(image_path)])[0]
    img_vec = np.array(img_vec).tolist()

    results = qdrant_client.query_points(
        collection_name=COLLECTION_NAME,
        query=img_vec,
        limit=k,
        with_payload=True
    )

    print(f"\nüñºÔ∏è Image search ‚Üí {image_path.name}")

    hits = []
    for i, r in enumerate(results.points, start=1):
        payload = r.payload or {}
        print(
            f"{i}. {payload.get('filename')} "
            f"(category={payload.get('category')}, score={r.score:.4f})"
        )
        hits.append({
            "id": r.id,
            "score": r.score,
            "filename": payload.get("filename"),
            "path": payload.get("path"),
            "category": payload.get("category"),
        })

    return hits
```
