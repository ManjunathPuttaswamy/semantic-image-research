```python
from langchain_experimental.open_clip import OpenCLIPEmbeddings

MODEL_NAME = "ViT-B-32"              # must be one of the built-ins
CHECKPOINT = "laion2b_s34b_b79k"     # the weights you want

embedder = OpenCLIPEmbeddings(
    model_name=MODEL_NAME,
    checkpoint=CHECKPOINT,
    device="cpu",
)
```

```python
import open_clip
print(open_clip.list_models())
```

```python
from PIL import Image

def embed_image(path: str):
    img = Image.open(path).convert("RGB")
    return embedder.embed_image(img)
```

```python
def embed_text(query: str):
    return embedder.embed_query(query)
```
